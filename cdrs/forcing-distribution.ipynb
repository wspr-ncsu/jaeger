```python
import numpy as np
import math
import matplotlib.pyplot as plt

def fx(_, k = 2.5, c = 1):
    k *= -1
    return c * math.pow(_, k)


x = np.array([1, 2, 3, 5, 6, 7, 8, 9, 10, 259,83,89,12,300,500,1029, 1200])
f = np.array([5326, 23, 234, 66, 234, 23, 62, 66, 3, 6, 2, 3, 2, 1, 4, 5, 3])
y = [fx(_) for _ in f]
y

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, y, linestyle='-', color='b', linewidth=1)

ax.set_title("$f(x) = x^{-2.5}$")
ax.set_xlabel("x")
ax.set_ylabel("f(x)")

plt.show()
```
